```python
import pandas as pd
import re

def clean_short_tweets(df):
    """Mark tweets with <10 words as non-duplicates"""
    
    # Mark short tweets as non-duplicates
    df.loc[df['word_count'] < 10, 'group'] = -1
    df.loc[df['word_count'] < 10, 'group_rank'] = -1
    df.loc[df['word_count'] < 10, 'is_duplicate'] = False
    
    return df

def mark_single_member_groups_as_non_duplicates(df):
    """Mark single-member duplicate groups as non-duplicates"""
    
    # Count members in each group
    group_sizes = df[df['group'] != -1]['group'].value_counts()
    
    # Identify groups with only one member
    single_member_groups = group_sizes[group_sizes == 1].index
    
    # Mark these groups as non-duplicates
    df.loc[df['group'].isin(single_member_groups), 'group'] = -1
    df.loc[df['group'].isin(single_member_groups), 'group_rank'] = -1
    df.loc[df['group'].isin(single_member_groups), 'is_duplicate'] = False
    
    print(f"Marked {len(single_member_groups)} single-member groups as non-duplicates")
    
    return df

# Apply cleaning
df_en = clean_short_tweets(df_en)
df_fa = clean_short_tweets(df_fa)

# Mark single-member groups as non-duplicates
df_en = mark_single_member_groups_as_non_duplicates(df_en)
df_fa = mark_single_member_groups_as_non_duplicates(df_fa)

# Now run your analysis with the cleaned data
# analyze_duplicate_groups(df_en, "English")
# analyze_duplicate_groups(df_fa, "Farsi")
```

```python
def plot_group_size_distribution(df, language):
    # Filter out non-duplicate groups (group == -1)
    duplicate_groups = df[df['group'] != -1]
    
    # Count ACTUAL group sizes (number of members per group)
    group_sizes = duplicate_groups['group'].value_counts()
    
    # Verify no single-member groups exist
    if 1 in group_sizes:
        print(f"Warning: Found {group_sizes[1]} single-member groups in {language} data")
        # Filter them out for visualization
        group_sizes = group_sizes[group_sizes > 1]
    
    # Create bins - now starting from 2
    bins = [2, 3, 4, 5, 10, 20, 50, 100, np.inf]
    labels = ['2', '3', '4', '5-9', '10-19', '20-49', '50-99', '100+']
    
    # Bin the actual group sizes (not value_counts of value_counts)
    binned = pd.cut(group_sizes, bins=bins, labels=labels, right=False)

    binned_counts = binned.value_counts().sort_index()
    
    # Plot
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=binned_counts.index, y=binned_counts.values, palette="viridis")
    plt.title(f'Distribution of Duplicate Group Sizes ({language})\n(Excluding Single-Member Groups)')
    plt.xlabel('Number of Duplicates in Group')
    plt.ylabel('Number of Groups')
    
    # Add value labels
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center',
                   xytext=(0, 5),
                   textcoords='offset points')
    
    plt.show()

# First verify your data
print("English data group size counts:")
print(df_en[df_en['group'] != -1]['group'].value_counts().value_counts())
print("\nPersian data group size counts:")
print(df_fa[df_fa['group'] != -1]['group'].value_counts().value_counts())

# Then plot
plot_group_size_distribution(df_en, 'English')
plot_group_size_distribution(df_fa, 'Persian')
```

```python
# Function to add ordered group size column
def add_ordered_group_column(df):
    # Get group sizes (excluding non-duplicates)
    group_sizes = df[df['group'] != -1].groupby('group').size()
    
    # Create mapping from group ID to ordered rank (1 = largest group)
    size_rank = group_sizes.rank(method='first', ascending=False).astype(int)
    
    # Create new column
    df['group_rank'] = df['group'].map(size_rank)
    
    # Non-duplicates get NaN (or you could set them to 0 or -1)
    df['group_rank'] = df['group_rank'].fillna(-1).astype(int)
    
    return df

# Apply to both datasets
df_en = add_ordered_group_column(df_en)
df_fa = add_ordered_group_column(df_fa)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter

def analyze_top_duplicate_groups(df, language, top_n=100):
    """Analyze top duplicate groups with temporal and content patterns."""
    print(f"\n{'='*50}\nAnalyzing top {top_n} duplicate groups - {language} tweets\n{'='*50}")
    
    # 1. Get top duplicate groups
    duplicate_groups = df[df['group_rank'] != -1]
    group_sizes = duplicate_groups['group_rank'].value_counts().head(top_n)
    
    # 2. Temporal analysis
    plt.figure(figsize=(14, 7))
    
    # Get creation dates for all tweets in top groups
    top_groups = group_sizes.index
    temporal_data = duplicate_groups[duplicate_groups['group_rank'].isin(top_groups)]
    
    # Plot daily frequency
    temporal_data['date'] = pd.to_datetime(temporal_data['created_at']).dt.date
    daily_counts = temporal_data.groupby(['group_rank', 'date']).size().unstack(0).fillna(0)
    
    # Plot top 10 groups for visibility
    daily_counts[group_sizes.index[:10]].plot(alpha=0.8, figsize=(9, 6), linewidth=2)
    #plt.title(f'Daily Frequency of Top 10 Duplicate Groups - {language}')
    plt.ylabel('Number of Tweets', fontsize=18)
    plt.xlabel('Date', fontsize=18)
    plt.xticks(fontsize=17, rotation=30, rotation_mode="anchor", ha='right')
    plt.yticks(fontsize=17)
    plt.rcParams['legend.title_fontsize'] = 18
    plt.legend(title='Near-Duplicate Group', loc='upper center', ncols=2, fontsize=16)
    plt.tight_layout()
    plt.savefig(f"figures/frequency_dup_{language}", dpi=300)
    plt.show()
    
    # 3. Group size distribution
    plt.figure(figsize=(12, 6))
    sns.histplot(group_sizes, bins=30, kde=True)
    plt.title(f'Distribution of Group Sizes (Top {top_n}) - {language}')
    plt.xlabel('Number of Duplicates in Group')
    plt.ylabel('Count')
    plt.show()
    
    # 4. Content analysis
    print(f"\nContent patterns in top {top_n} duplicate groups:")
    
    # Get representative samples
    samples = []
    for group_id in group_sizes.index[:20]:  # Show top 10 for brevity
        group_tweets = duplicate_groups[duplicate_groups['group_rank'] == group_id]
        
        # Get most representative tweet (median length)
        rep_tweet = group_tweets.iloc[(group_tweets['tweet_clean'].str.len() - 
                                     group_tweets['tweet_clean'].str.len().median()).abs().argsort()[:1]]
        
        samples.append({
            'group_id': group_id,
            'size': len(group_tweets),
            'example': rep_tweet['tweet_clean'].values[0][:200] + ('...' if len(rep_tweet['tweet_clean'].values[0]) > 200 else ''),
            'first_date': group_tweets['created_at'].min(),
            'last_date': group_tweets['created_at'].max()
        })
    
    # Display as a table
    samples_df = pd.DataFrame(samples)
    print(samples_df.to_string(index=False))
    
    # 5. Hashtag analysis (if available)
    if 'hashtags' in df.columns:
        top_group_tweets = duplicate_groups[duplicate_groups['group_rank'].isin(top_groups)]
        hashtags = top_group_tweets['hashtags'].explode().value_counts().head(10)
        
        plt.figure(figsize=(10, 5))
        sns.barplot(y=hashtags.index, x=hashtags.values, palette="rocket")
        plt.title(f'Top Hashtags in Duplicate Groups - {language}')
        plt.xlabel('Count')
        plt.show()
    
    return group_sizes

# Analyze English tweets
top_en_groups = analyze_top_duplicate_groups(df_en, "English")

# Analyze Persian tweets
top_fa_groups = analyze_top_duplicate_groups(df_fa, "Persian")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

def analyze_random_duplicate_groups(df, language, sample_size=100, examples_per_group=3):
    """Analyze random duplicate groups with multiple examples from each group."""
    print(f"\n{'='*60}\nAnalyzing {sample_size} random duplicate groups - {language} tweets\n{'='*60}")
    
    # Filter to only include duplicate tweets (groups with rank != -1)
    duplicate_groups = df[df['group_rank'] != -1].copy()
    
    # Get unique group IDs
    unique_groups = duplicate_groups['group_rank'].unique()
    
    # Randomly select groups
    if len(unique_groups) > sample_size:
        selected_groups = random.sample(list(unique_groups), sample_size)
    else:
        selected_groups = unique_groups
        print(f"Only {len(unique_groups)} groups available, analyzing all of them")
    
    # Collect examples
    samples = []
    for group_id in selected_groups:
        group_tweets = duplicate_groups[duplicate_groups['group_rank'] == group_id]
        
        # If there are fewer tweets than requested examples, take what's available
        n_examples = min(examples_per_group, len(group_tweets))
        
        # Randomly select examples from the group
        example_tweets = group_tweets.sample(n_examples, random_state=42)
        
        for idx, (_, tweet) in enumerate(example_tweets.iterrows()):
            samples.append({
                'group_id': group_id,
                'example_num': idx + 1,
                'tweet_text': tweet['tweet'][:500] + ('...' if len(tweet['tweet']) > 500 else ''),
                'created_at': tweet['created_at'],
                'group_size': len(group_tweets)
            })
    
    # Create DataFrame for better visualization
    samples_df = pd.DataFrame(samples)
    
    # Display the results
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_colwidth', 100)
    
    print(f"\nSample tweets from {len(selected_groups)} random duplicate groups:")
    print(samples_df.to_string(index=False))
    
    return samples_df

# Set random seed for reproducibility
random.seed(42)
```

```python
import pandas as pd

def get_top_group_examples(df, language, top_n=20):
    """Return one example tweet per group for the top N duplicate groups."""
    duplicate_groups = df[df['group_rank'] != -1]
    group_sizes = duplicate_groups['group_rank'].value_counts().head(top_n)
    
    samples = []
    for group_id in group_sizes.index:
        group = duplicate_groups[duplicate_groups['group_rank'] == group_id]
        rep_tweet = group.iloc[(group['tweet'].str.len() - group['tweet_clean'].str.len().median()).abs().argsort()[:1]]
        samples.append({
            'group': group_id,
            'size': len(group),
            'accounts': group['author id'].nunique(),
            'example': rep_tweet['tweet'].values[0].replace('\n', ' ')
        })
    return pd.DataFrame(samples)

# Get tables
top_en_df = get_top_group_examples(df_en, "English")
top_fa_df = get_top_group_examples(df_fa, "Persian")

# LaTeX table template
def make_latex_table(df, caption, label, is_persian=False):
    rows = []
    for i, row in df.iterrows():
        example = row['example']
        if is_persian:
            example = example.replace('_', r'\_')  # escape underscores
        rows.append(f"{i+1} & {row['size']} & {row['accounts']} & {example} \\\\")

    latex = rf"""
\begin{{table*}}[t]
\caption{{{caption}}}
\renewcommand{{\arraystretch}}{{1.2}}
\setlength{{\tabcolsep}}{{5pt}}
\small
\begin{{tabularx}}{{\textwidth}}{{>{{\centering\arraybackslash}}p{{0.05\textwidth}} >{{\centering\arraybackslash}}p{{0.10\textwidth}} >{{\centering\arraybackslash}}p{{0.10\textwidth}} >{{\raggedright\arraybackslash}}p{{0.65\textwidth}}}}
\hline
\textbf{{Group}} & \textbf{{Size}} & \textbf{{Accounts}} & \textbf{{Example Tweet}} \\
\hline
{chr(10).join(rows)}
\hline
\label{{{label}}}
\end{{tabularx}}
\end{{table*}}
"""
    return latex

# Print LaTeX code
english_latex = make_latex_table(top_en_df, "Top English Duplicate Tweet Groups with Examples", "tableenglishdup")
persian_latex = make_latex_table(top_fa_df, "Top Persian Duplicate Tweet Groups with Examples and Translation", "tablepersiandup", is_persian=True)

# Save or print
with open("english_duplicate_table.tex", "w", encoding="utf-8") as f:
    f.write(english_latex)

with open("persian_duplicate_table.tex", "w", encoding="utf-8") as f:
    f.write(persian_latex)

print("LaTeX tables saved to disk.")
```

```python
import random

def analyze_duplicate_groups(df, dataset_name, group_col="group_rank"):
    # Set random seed for reproducibility
    random.seed(42)
    
    # Create is_duplicate flag
    df["is_duplicate"] = df[group_col] != -1
    
    # Group by duplicate group and count frequency
    group_counts = df.groupby(group_col).agg(
        count=('author id', 'size'),
        unique_authors=('author id', 'nunique')
    ).reset_index()
    
    # Sort by count in descending order
    group_counts = group_counts.sort_values(by="count", ascending=False)
    
    # Separate duplicates and non-duplicates
    duplicate_groups = group_counts[group_counts[group_col] != -1]
    non_duplicates = group_counts[group_counts[group_col] == -1]
    
    # Get the top 5 duplicate groups for LaTeX table
    top_5_groups = duplicate_groups.head(5)
    
    # Generate LaTeX table
    latex_table = []
    latex_table.append(r"\begin{tabular}{|c|c|l|}")
    latex_table.append(r"\hline")
    latex_table.append(r"Group Size & Unique Authors & Example Tweets \\")
    latex_table.append(r"\hline")
    
    for _, row in top_5_groups.iterrows():
        group_id = row[group_col]
        group_df = df[df[group_col] == group_id]
        
        # Get random sample (using random.sample for true randomness with seed)
        if len(group_df) >= 3:
            random_indices = random.sample(range(len(group_df)), 3)
            group_tweets = group_df.iloc[random_indices]
        else:
            group_tweets = group_df
        
        # Add group info row
        latex_table.append(f"{row['count']} & {row['unique_authors']} & \\\\")
        
        # Add tweet examples
        for i, (_, tweet_row) in enumerate(group_tweets.iterrows(), 1):
            tweet_text = tweet_row['tweet'].replace('&', r'\&').replace('%', r'\%').replace('_', r'\_')
            if dataset_name == "Farsi":
                tweet_text = r"\textfarsi{" + tweet_text + "}"
            latex_table.append(f" & & {i}. {tweet_text} \\\\")
        
        latex_table.append(r"\hline")
    
    latex_table.append(r"\end{tabular}")
    
    # Save LaTeX table to file
    with open(f"{dataset_name.lower()}_duplicate_groups.tex", "w", encoding='utf-8') as f:
        f.write("\n".join(latex_table))
    
    print(f"\nLaTeX table for top 5 duplicate groups in {dataset_name} dataset saved to {dataset_name.lower()}_duplicate_groups.tex")
    
    # Plotting
    plot_data = duplicate_groups.head(20).copy()
    plot_data[group_col] = plot_data[group_col].astype(str)  # Force categorical for horizontal bars

    
    
    plt.figure(figsize=(10, 7))
    plt.ticklabel_format(style='plain', axis='x')
    sns.barplot(
        data=plot_data,
        y=group_col,         # Y = group name (horizontal bar orientation)
        x="count",           # X = number of tweets
        color="mediumseagreen"
    )
    
    plt.xlabel("Number of Tweets", fontsize=18)
    plt.xticks(fontsize=17)
    plt.yticks(fontsize=18)
    plt.ylabel("Near-Duplicate Group", fontsize=18)
    #plt.title(f"Top 50 Duplicate Groups - {dataset_name} Dataset")
    plt.tight_layout()
    plt.savefig(f"figures/{dataset_name.lower()}_top_duplicate_groups.png", dpi=300)
    plt.show()
    
    # Print examples to console
    if not top_5_groups.empty:
        print(f"\nRandom Examples of the 5 Most Frequent Duplicate Groups ({dataset_name}):")
        
        for _, row in top_5_groups.iterrows():
            group_id = row[group_col]
            group_df = df[df[group_col] == group_id]
            
            if len(group_df) >= 3:
                random_indices = random.sample(range(len(group_df)), 3)
                group_tweets = group_df.iloc[random_indices]
            else:
                group_tweets = group_df
            
            print(f"\nGroup {group_id} (Size: {row['count']}, Unique Authors: {row['unique_authors']}):")
            for i, (_, tweet_row) in enumerate(group_tweets.iterrows(), 1):
                print(f"  Example {i}: {tweet_row['tweet']}")

analyze_duplicate_groups(df_en, "English")
analyze_duplicate_groups(df_fa, "Persian")
```

```python
import pandas as pd

def compute_author_stats(df, label):
    df = df.copy()
    df['created_at'] = pd.to_datetime(df['created_at'])
    df['date'] = df['created_at'].dt.date
    df['hour'] = df['created_at'].dt.hour
    df = df.sort_values(['author_username', 'created_at'])
    df['time_diff'] = df.groupby('author_username')['created_at'].diff().dt.total_seconds()
    
    df['is_duplicate'] = df['group_rank'].notna() & (df['group_rank'] != -1)

    stats = df.groupby('author_username').agg({
        'created_at': ['count', lambda x: x.dt.date.nunique()],
        'time_diff': ['mean', 'median'],
        'is_duplicate': 'sum',
        'hour': lambda x: ((x < 6) | (x > 22)).sum()
    })

    stats.columns = [
        'Total Tweets',
        'Active Days',
        'Avg Time Between Tweets (s)',
        'Median Time Between Tweets (s)',
        'Repeated Tweets',
        'Night Tweets'
    ]

    stats['Tweets per Day'] = stats['Total Tweets'] / stats['Active Days']
    stats['Repeated Tweet %'] = 100 * stats['Repeated Tweets'] / stats['Total Tweets']
    stats['Night Tweet %'] = 100 * stats['Night Tweets'] / stats['Total Tweets']

    # Flag behavior
    stats['Flag: High Volume'] = stats['Tweets per Day'] > 100
    stats['Flag: High Repetition'] = stats['Repeated Tweet %'] > 80
    stats['Flag: Night Activity'] = stats['Night Tweet %'] > 40
    stats['Flag: Fast Posting'] = stats['Median Time Between Tweets (s)'] < 60

    stats['Likely Automated'] = stats[
        ['Flag: High Volume', 'Flag: High Repetition', 'Flag: Night Activity', 'Flag: Fast Posting']
    ].sum(axis=1) >= 2

    stats['group_label'] = label
    return stats.reset_index()

# # Prepare control and duplicate datasets
# df_control = df_en[df_en['group'] == -1]
# df_duplicate = df_en[df_en['group'] != -1]

# Identify authors in each group
authors_control = set(df_en[df_en['group_rank'] == -1]['author_username'])
authors_duplicate = set(df_en[df_en['group_rank'] != -1]['author_username'])

# Remove overlapping authors
exclusive_control_authors = authors_control - authors_duplicate
exclusive_duplicate_authors = authors_duplicate - authors_control

# Filter the DataFrame for mutually exclusive groups
df_control = df_en[(df_en['group_rank'] == -1) & (df_en['author_username'].isin(exclusive_control_authors))]
df_duplicate = df_en[(df_en['group_rank'] != -1) & (df_en['author_username'].isin(exclusive_duplicate_authors))]

# Compute author-level stats for each group
control_stats = compute_author_stats(df_control, 'control')
duplicate_stats = compute_author_stats(df_duplicate, 'duplicate')

# Combine
combined = pd.concat([control_stats, duplicate_stats], ignore_index=True)

# Summarize at group level
group_summary = combined.groupby('group_label').agg({
    'Total Tweets': 'sum',
    'Active Days': 'sum',
    'Tweets per Day': 'mean',
    'Repeated Tweet %': 'mean',
    'Night Tweet %': 'mean',
    'Median Time Between Tweets (s)': 'median',
    'Likely Automated': 'mean'
})

group_summary['Likely Automated %'] = 100 * group_summary['Likely Automated']
group_summary = group_summary.drop(columns='Likely Automated')

# Show summary
print(group_summary)
```

```python
import pandas as pd
import re

# Function to extract mentions from a tweet and remove the '@' sign
def extract_mentions(text):
    return [mention[1:] for mention in re.findall(r'@\w+', text)]  # Remove '@' sign

# Extract mentions for all tweets
df['mentions'] = df['tweet'].apply(extract_mentions)

# Separate duplicates and non-duplicates
duplicates = df[df['is_duplicate'] == 1]
non_duplicates = df[df['is_duplicate'] == 0]

# Get unique authors in each group
duplicate_authors = set(duplicates['author_username'])
non_duplicate_authors = set(non_duplicates['author_username'])

# Function to count mentions within the same group
def count_mentions_within_group(tweets, authors_in_group):
    within_group_mentions = 0
    total_mentions = 0
    
    for mentions in tweets['mentions']:
        for mention in mentions:
            total_mentions += 1
            if mention in authors_in_group:
                within_group_mentions += 1
    
    return within_group_mentions, total_mentions

# Count mentions within the same group for duplicates
duplicate_within_group, duplicate_total = count_mentions_within_group(duplicates, duplicate_authors)

# Count mentions within the same group for non-duplicates
non_duplicate_within_group, non_duplicate_total = count_mentions_within_group(non_duplicates, non_duplicate_authors)

# Calculate percentages
duplicate_percentage = (duplicate_within_group / duplicate_total) * 100 if duplicate_total > 0 else 0
non_duplicate_percentage = (non_duplicate_within_group / non_duplicate_total) * 100 if non_duplicate_total > 0 else 0

# Print results
print(f"Duplicates: {duplicate_within_group} / {duplicate_total} mentions are within the same group ({duplicate_percentage:.2f}%)")
print(f"Non-Duplicates: {non_duplicate_within_group} / {non_duplicate_total} mentions are within the same group ({non_duplicate_percentage:.2f}%)")
```
